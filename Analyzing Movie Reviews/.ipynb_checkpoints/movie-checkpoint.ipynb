## Differences between Metacritic vs. Fandango
### - Analysis
* The distribution of the Metacritic_norm_round is more unifrom and wider
* The districution of the Fandango_Stars is sharper
* Fandango shows higher score on average
* The correlation between Metacritic_norm_round and Fandango_Stars is not strong (slightly positive correlation)

```python
plt.hist(movies["Metacritic_norm_round"],bins="auto", label="Metacritic_norm_round")
plt.hist(movies["Fandango_Stars"], bins="auto", label="Fandango_Stars")
plt.legend()
plt.show()
```

```python
mean_F = movies["Fandango_Stars"].mean()
mean_M = movies["Metacritic_norm_round"].mean()
std_F = np.std(movies["Fandango_Stars"])
std_M = np.std(movies["Metacritic_norm_round"])
print("For Fandango_Stars: mean = {0:.2f}; std = {1:.2f}".format(mean_F, std_F))
print("For Metacritic_norm_round: mean = {0:.2f}; std = {1:.2f}".format(mean_M, std_M))
```

```python
plt.scatter(movies["Fandango_Stars"], movies["Metacritic_norm_round"])
```

* Belows are the top 5 films with the largest difference between Metacritic_norm_round and Fandango_Stars

```python
movies["fm_diff"] = movies["Metacritic_norm_round"] - movies["Fandango_Stars"]
movies["fm_diff"] = abs(movies["fm_diff"])
temp = movies["fm_diff"].sort_values(ascending = False)
index_list = list(temp[0:5].index)
for i in index_list:
    top_5 = movies["FILM"].loc[i]
    diff = movies["fm_diff"].loc[i]
    print("Index: {0}; Film: {1}; fm_diff: {2}".format(i,top_5,diff))
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
print("r value: {:.2f}".format(r_value))
```

```python
def prediction(x):
    predict = slope*x + intercept
    return predict

print("Prediction funciton: y = {0:.2f}x + {1:.2f}".format(slope, intercept))
print("Predicted value (x = 3): {:.2f}".format(prediction(3.0)))
```

```python
sns.residplot(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
```

```python
predict_1 = prediction(1.0)
predict_5 = prediction(5.0)

plt.scatter(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
x = movies["Metacritic_norm_round"]
plt.plot([1.0, 5.0], [predict_1, predict_5])
plt.xlim(1, 5)
plt.xlabel("Metacritic_norm_round")
plt.ylabel("Fandango_Stars")
```
