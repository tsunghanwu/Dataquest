## Dataset Overview

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")
star_wars.head(10)
```

```python
print(star_wars.columns)
```

## Data cleaning

### - Filter out the data without ID

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

### - Convert data into Booleans for analysis

```python
yes_no = {"Yes":True, "No":False}
selected = ["Have you seen any of the 6 films in the Star Wars franchise?", "Do you consider yourself to be a fan of the Star Wars film franchise?"]
for i in selected:
    star_wars[i] = star_wars[i].map(yes_no)

star_wars.head()
```

```python
import numpy as np
seen = {"Star Wars: Episode I  The Phantom Menace": True, "Star Wars: Episode II  Attack of the Clones": True, 
        "Star Wars: Episode III  Revenge of the Sith": True, "Star Wars: Episode IV  A New Hope": True, 
        "Star Wars: Episode V The Empire Strikes Back": True, "Star Wars: Episode VI Return of the Jedi": True,
        np.nan: False}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen)

star_wars.head()
```

### - Rename columns names

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"})
```

```python
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

ranking = {}
number = 1
for col in star_wars.columns[9:15]:
    ranking[col] = "ranking_"+ str(number)
    number += 1

star_wars = star_wars.rename(columns=ranking)
```

## Analysis

### - Highest ranked movie
* Base on the mean values and bar chart, the earlier movies show higher ranking

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
index = np.linspace(0,10,6) + 0.5
bar_width = 0.5
ax.bar(index, star_wars[star_wars.columns[9:15]].mean(), bar_width)
plt.xticks(index, star_wars.columns[9:15], rotation=60)
```

### - View Counts
* Based on the summation and bar chart, the latter movies show higher view counts (but their rankings are lower)
* It suggests that the earlier movies are more satisfying

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
fig_2, ax_2 = plt.subplots()
ax_2.bar(index, star_wars[star_wars.columns[3:9]].sum(), bar_width)
plt.xticks(index, star_wars.columns[3:9], rotation = 60)
```

### - The difference between genders

* Both male and female audience prefer the earlier movies
* The earlier movies are more satisfying for both male and female audience
* Compared with male audience, the earlier movies (ep1-3) are more satisfying for female audience

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
fig_3, ax_3 = plt.subplots()
ax_3.bar(index, males[males.columns[9:15]].mean(), bar_width, label = "Males")
ax_3.bar(index + bar_width, females[females.columns[9:15]].mean(), bar_width, label = "Females")
plt.xticks(index, males.columns[9:15], rotation=60)
plt.legend()
```

```python
fig_4, ax_4 = plt.subplots()
ax_4.bar(index, males[males.columns[3:9]].sum(), bar_width, label = "Males")
ax_4.bar(index + bar_width, females[males.columns[3:9]].sum(), bar_width, label = "Females")
plt.xticks(index, males.columns[3:9], rotation = 60)
plt.legend()
```
