## Dataset overview

```python
import pandas as pd
import numpy as np

columns = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 
           'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv("imports-85.data", names=columns)
cars.head()
```

## Data cleaning
### - Replace meaningless characters

```python
cars["normalized-losses"] = cars["normalized-losses"].replace("?", np.nan)
cars.head()
```

```python
cars.info()
```

```python
numeric_value_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[numeric_value_cols]
numeric_cars.head()
```

```python
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars.head()
```

### - Handle NA values

```python
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

## Feature scaling
### - Data normalization

```python
numeric_cars_price = numeric_cars["price"]
numeric_cars = (numeric_cars - numeric_cars.min()) / (numeric_cars.max() - numeric_cars.min())
numeric_cars["price"] = numeric_cars_price
numeric_cars
```

##  Build an univariate KNN model (with default k = 5)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    last_train_row = int(len(shuffled_df)/2)
    train_df = shuffled_df.iloc[:last_train_row]
    test_df = shuffled_df.iloc[last_train_row:]
    
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]], train_df["price"])
    predictions = knn.predict(test_df[[train_col]])
    
    mse = mean_squared_error(test_df["price"], predictions)
    rmse = mse**(1/2)
    return rmse
```

### - Find the train column that shows minimum RMSE
* Horsepower shows the minimum RMSE

```python
rmse_dict = {}
train_columns = numeric_cars.columns.drop("price")

for column in train_columns:
    rmse_dict[column] = knn_train_test(column, "price", numeric_cars)
rmse_dict
```

```python
rmse_series = pd.Series(rmse_dict)
rmse_series = rmse_series.sort_values()
rmse_series
```

## Build univariate KNN models (with k = 1,3,5,7,9)

```python
def knn_train_test2(train_col, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    last_train_row = int(len(shuffled_df)/2)
    train_df = shuffled_df.iloc[:last_train_row]
    test_df = shuffled_df.iloc[last_train_row:]
    
    k_values = [1,3,5,7,9]
    k_rmses = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]], train_df["price"])
        predictions = knn.predict(test_df[[train_col]])
        
        mse = mean_squared_error(test_df["price"], predictions)
        rmse = mse**(1/2)
        k_rmses[k] = rmse
        
    return k_rmses
```

```python
k_rmse_dict = {}
train_columns = numeric_cars.columns.drop("price")
for column in train_columns:
    k_rmse_dict[column] = knn_train_test2(column, "price", numeric_cars)

k_rmse_dict
```

### - Data visualization for k-value vs. RMSE 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#line_color = sns.color_palette("hls", len(k_rmse_dict)+1)
#count = 0
for column, value in k_rmse_dict.items():
    k_value = list(value.keys())
    k_value_rmse = list(value.values())
    count += 1
    plt.plot(k_value,k_value_rmse,"-o", label=column)
    plt.xlabel("k_value")
    plt.ylabel("RMSE")
    
plt.xticks(range(10))
plt.xlim(0,16)
plt.legend()
```

### - Find the column with minimum average RMSE
* Horsepower shows minimum average RMSE

```python
feature_avg_rmse = {}
for key, value in k_rmse_dict.items():
    avg_rmse = np.mean(list(value.values()))
    feature_avg_rmse[key] = avg_rmse
avg_rmse_series = pd.Series(feature_avg_rmse)
avg_rmse_series = avg_rmse_series.sort_values()
avg_rmse_series
```

## Build a multivariate KNN model (with k=5)

```python
def knn_train_test3(train_list, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    last_train_row = int(len(shuffled_df)/2)
    train_df = shuffled_df.iloc[:last_train_row]
    test_df = shuffled_df.iloc[last_train_row:]
    
    k_values = [5]
    k_rmses = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_list], train_df["price"])
        predictions = knn.predict(test_df[train_list])
        
        mse = mean_squared_error(test_df["price"], predictions)
        rmse = mse**(1/2)
        k_rmses[k] = rmse
        
    return k_rmses
```

### - Find the minimum RMSE for different feature combinations
* Four_best_features combination (horsepower, width, curb-weight, city-mpg) shows minimum RMSE

```python
k_rmse_results = {}

two_best_features = ['horsepower', 'width']
rmse_val = knn_train_test3(two_best_features, 'price', numeric_cars)
k_rmse_results["two best features"] = rmse_val

three_best_features = ['horsepower', 'width', 'curb-weight']
rmse_val = knn_train_test3(three_best_features, 'price', numeric_cars)
k_rmse_results["three best features"] = rmse_val

four_best_features = ['horsepower', 'width', 'curb-weight', 'city-mpg']
rmse_val = knn_train_test3(four_best_features, 'price', numeric_cars)
k_rmse_results["four best features"] = rmse_val

five_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg']
rmse_val = knn_train_test3(five_best_features, 'price', numeric_cars)
k_rmse_results["five best features"] = rmse_val

six_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg', 'length']
rmse_val = knn_train_test3(six_best_features, 'price', numeric_cars)
k_rmse_results["six best features"] = rmse_val

k_rmse_results
```

## Build multivariate KNN models (with k = 1 to 26)

```python
def knn_train_test4(train_list, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    last_train_row = int(len(shuffled_df)/2)
    train_df = shuffled_df.iloc[:last_train_row]
    test_df = shuffled_df.iloc[last_train_row:]
    
    k_values = [i for i in range(1,26,1)]
    k_rmses = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_list], train_df["price"])
        predictions = knn.predict(test_df[train_list])
        
        mse = mean_squared_error(test_df["price"], predictions)
        rmse = mse**(1/2)
        k_rmses[k] = rmse
        
    return k_rmses
```

### - Hyperparameter tuning
* For lower k-value (k <= 10), four_best_features combination shows lower RMSE
* For higher k-value (k > 10), three_best_features combination shows lower RMSE
* In general, lower k-value shows lower RMSE

```python
k_rmse_results = {}

three_best_features = ['horsepower', 'width', 'curb-weight']
rmse_val = knn_train_test4(three_best_features, 'price', numeric_cars)
k_rmse_results["three best features"] = rmse_val

four_best_features = ['horsepower', 'width', 'curb-weight', 'city-mpg']
rmse_val = knn_train_test4(four_best_features, 'price', numeric_cars)
k_rmse_results["four best features"] = rmse_val

five_best_features = ['horsepower', 'width', 'curb-weight' , 'city-mpg' , 'highway-mpg']
rmse_val = knn_train_test4(five_best_features, 'price', numeric_cars)
k_rmse_results["five best features"] = rmse_val

k_rmse_results
```

```python
for column, value in k_rmse_results.items():
    k_value = list(value.keys())
    k_value_rmse = list(value.values())
    plt.plot(k_value, k_value_rmse, "-o", label=column)
    plt.xlabel("k_value")
    plt.ylabel("RMSE")
plt.legend()
```
